### 1. Poisson $\rightarrow$ Normal Approximation

#### In class we showed that the Poisson distribution represents the limiting behavior of the binomial distribution for large $n$, small $p$, and constant $np$. When to use the Poisson distribution as an approximation for the binomial will always depend on the accuracy needed for your specific problem, but a reasonably good criterion is when $n < 100$, $p < 0.1$, and $np < 20$ . Similarly, by varying p and n, we also showed that the normal distribution is a pretty good approximation to the binomial when $np > 5$ and $n(1-p) > 5$. The Poisson distribution also approaches the normal distribution when $\lambda$  is large. In this limit, the mean and variance of the normal distribution are both $\lambda$ :

$$ P(k) = \frac{e^{\lambda} \lambda^k}{k!} \approx \frac{1}{\sqrt{2 \pi \lambda}} e^{-\frac{1}{2} \frac{(k - \lambda)^2}{k}}$$

### 2. Giant planet eccentricities

#### Giant planets have been found to have a wide range of orbital eccentricities from e=0 (circular orbits) to $e>0.9$. The distribution of giant planet eccentricities provides clues about how common past dynamical encounters were with binary companions or other planets in the system because most planets are thought to form with near-circular orbits. Kipping (2013, MNRAS, 434, L51) found that the ensemble of eccentricities of giant planets can be well fit with a Beta distribution with exponents $a=0.867$ and $b=3.03$: 

$$ P(e|a,b) = \frac{\Gamma (a+b)}{\Gamma(a) \Gamma(b)} e^{a - 1}(1-e)^{b-1}$$

#### where $\Gamma$ is the Gamma function $(\Gamma(n) = (n - 1)!)$.

### 3. The Central Limit Theorem

#### When independent random variables are added together they converge to a normal distribution. This is a feature of the Central Limit Theorem and explains why the Gaussian distribution is seen so frequently in astronomy (and all of science). It provides the theoretical foundation for how to improve the accuracy of a result with repeated measurements. For any arbitrary distribution with mean $\mu$ and variance $\sigma^2$ 2, the mean of N random values drawn from that distribution will approximately follow a normal distribution with a mean of $\mu$ and variance of  $\sigma^2/N$. The exponential distribution has the following form:
#### The exponential distribution has the following form:

$$P(x|\lambda) = \begin{array}{cc}
  \{ & 
    \begin{array}{cc}
      0 & x < 0 \\
      \lambda e^{-\lambda x} & x \geq 0 \\
    \end{array}
\end{array}$$

#### The mean value of the exponential distribution is $1/\lambda$  and the variance is $1/\lambda^2$. For the following questions assume  $\lambda=2$.

### 4. Sample Statistics
#### Draw random values with a sample size of N from a standard normal distribution with $\mu =0$ and  $\sigma=1$.

#### (4a) [Include code] Plot the sample mean and unbiased sample variance (with N - 1 in the denominator) as a function of N from N=2 to N=100. Do $\bar{x}$ and $s^2$ approach $\mu$ and $\sigma^2$?

```python
#This is the code for all 3 parts of 4. 

sam_mean_4a = []     #sample mean
sam_var = []  #Unbiased sample variance
sam_var_bias = []  #Biased sample variance 
unbias_sd = []   #Unbiased std dev

sum = 0
for N in range(2,101):
    mu, sigma = 0, 1
    sample = np.random.normal(mu, sigma, N)
    x_sum = np.sum(sample)
    N_const = 1/(N-1)
    N_const_bias = 1/N
    #sample mean
    sam_mean_4a.append(x_sum*N_const)
    #Unbiased sample variance
    var_sum = np.sum((sample - (x_sum*N_const))**2)
    sam_var.append(var_sum*N_const)
    #Biased sample variance 
    var_sum_bias = np.sum((sample - (x_sum*N_const_bias))**2)
    sam_var_bias.append(var_sum_bias*N_const_bias)
    #Unbiased std dev 
    unbias_sd.append(np.sqrt(var_sum*N_const))
    
```

```python
plt.scatter(range(2,101), sam_mean_4a, color ="purple")


plt.xlabel('N', fontsize=12)
plt.ylabel("Sample Mean", fontsize =14)
#plt.legend(loc='lower right', borderaxespad=0., fontsize = 12)

plt.ylim(-1.5,1.5)
ax = plt.gca()
ax.grid()
ax.tick_params(which="both", 
               bottom=True, top=True, left=True, right=True, 
               direction = "in", 
               grid_color = "gray", grid_alpha = 0.3)
```

```python
plt.scatter(range(2,101), sam_var, color = "purple")

plt.xlabel('N', fontsize=12)
plt.ylabel("Unbiased Sample Variance", fontsize =14)
#plt.legend(loc='lower right', borderaxespad=0., fontsize = 12)

plt.ylim(-1,3)
ax = plt.gca()
ax.grid()
ax.tick_params(which="both", 
               bottom=True, top=True, left=True, right=True, 
               direction = "in", 
               grid_color = "gray", grid_alpha = 0.3)
```

Conclusion: Yes, $\bar{x}$ approaches $\mu$ and $s^2$ approaches $\sigma^2$ !

#### (4c) [Include code] Plot the unbiased sample standard deviation (the standard deviation of the unbiased sample variance) as a function of N from N=2 to N=100. Is $\sqrt{s^2}$ a good estimate of $\sigma$ ? Why does $\sqrt{s^2}$ systematically underestimate $\sigma$ ?

```python
plt.scatter(range(2,101), unbias_sd, color = "purple")

plt.xlabel('N', fontsize=14)
plt.ylabel("Unbiased Sample SD", fontsize =14)
#plt.legend(loc='lower right', borderaxespad=0., fontsize = 12)

plt.ylim(-1,3)
ax = plt.gca()
ax.grid()
ax.tick_params(which="both", 
               bottom=True, top=True, left=True, right=True, 
               direction = "in", 
               grid_color = "gray", grid_alpha = 0.3)
```

Conclusion: $\sqrt{s^2}$ can be a reasonable estimate of $\sigma$ for high N. But this systematically underestimates $\sigma$ because there are N degrees of freedom when calculating $\bar{x}$ but for calculating $s^2$ there is just $N-1$.
